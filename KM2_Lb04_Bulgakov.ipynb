```python
def regular_polygon_Relo(n:int=9, center:tuple=(0,0), r:int=14, N:int=100) -> list:

    assert n > 2 and n % 2 == 1,'n is more than 2 and odd'
    assert type(center) == tuple and len(center) == 2, 'center is tuple of 2 elements'
    assert r > 0, 'radius r should be positive'
    assert N > 0, 'number of points should be positive'
    l:float = 2*r*np.sin(np.pi/(2*n))
    R:float = l/(2*np.sin(np.pi/n))
    t:list = np.arange(0,2*np.pi, 2*np.pi/n)
    vertices:list = center + R*np.transpose([np.cos(t),np.sin(t)])
    alpha:float = 2*np.pi/n
    beta:float = alpha/2
    angle:list = np.linspace(-beta/2, beta/2,N)
    side:list = [vertices[i] + r*np.transpose([np.cos(np.pi+angle+i*alpha), np.sin(np.pi+angle+i*alpha)]) for i in range(n)]
    sides:list = np.concatenate(side)
    return sides
```

```python
regular_polygon_Relo(9, (0,0), 14, 100)
```
